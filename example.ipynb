```python
# Imports
import sensemakr
import sensitivity_stats
import bias_functions
import ovb_bounds
import ovb_plots
import statsmodels.api as sm
import numpy as np
import pandas as pd
```

```python
# Some sample data
a = [-2.38092296,  0.62183556,  0.02550352,  1.42789457, -1.94621868, -0.61530634, 1.03725401,  0.56475839, -0.37657177, -1.10448260]
b = [0.63172540, 1.47700758, 0.45603515, 0.89147827, -0.20224887, 0.24311158, 0.60697111, -1.35162322, -1.13587531, 0.06992495]
c = [0.8019378, 0.5549415, 0.3406929, -1.4490734, -0.6485534, -1.4420249, -0.8218735, 1.9395321, -1.2180532, -0.6232199]
y = [0.1626037, -2.4405759, -1.4550886, 0.4962169, -0.4558373, 1.2204772, .2503206, -1.4071841, -0.9135852, 1.4494329]
```

```python
# Fit a statsmodels OLS object to get a statsmodels OLSResults object
X = pd.DataFrame({'a': a, 'b': b, 'c': c})
X = sm.add_constant(X)
reg_model = sm.OLS(y,X)
model = reg_model.fit()
```

```python
# Define parameters for sensemakr
treatment='a'
q = 1.0
alpha = 0.25
reduce = True
benchmark_covariates=['b', ['b', 'c']]
kd = [0.1, 0.2]
ky = kd
```

```python
# Create a sensemakr object and print summary of results
s = sensemakr.Sensemakr(model, treatment, q=q, alpha=alpha, reduce=reduce, benchmark_covariates=benchmark_covariates, kd=kd)
s.summary()
```

```python
# Make a contour plot for the estimate
ovb_plots.ovb_contour_plot(sense_obj=s, sensitivity_of='estimate')
```
